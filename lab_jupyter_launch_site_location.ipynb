```python
# Task 2: Mark the success/failed launches for each site on the map
# Color mapping for launch outcome
def outcome_color(outcome):
    if outcome == 1:
        return 'green'  # success
    else:
        return 'red'    # failure

# Add each launch as a circle
for i, row in spacex_df.iterrows():
    folium.Circle(
        location=[row['Lat'], row['Long']],
        radius=500,  # smaller radius for individual launches
        color=outcome_color(row['class']),
        fill=True,
        fill_opacity=0.7,
    ).add_child(folium.Popup(f"{row['Launch Site']}Outcome: {'Success' if row['class']==1 else 'Failure'}")
    ).add_to(site_map)

# Add labels for launch sites (one per site)
for i, row in launch_sites_df.iterrows():
    folium.map.Marker(
        [row['Lat'], row['Long']],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:#0000FF;"><b>{row["Launch Site"]}</b></div>'
        )
    ).add_to(site_map)

site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np
# Haversine formula to calculate distance (km) between two lat/lon points
def haversine(coord1, coord2):
    lat1, lon1 = np.radians(coord1)
    lat2, lon2 = np.radians(coord2)
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    R = 6371  # Earth radius in kilometers
    return R * c

# Example: Launch site (CCAFS LC-40)
launch_site = [28.562302, -80.577356]

# Example proximities
city = [28.538336, -81.379234]       # Orlando, FL
railway = [28.563197, -80.567123]    # Example railway coord
highway = [28.5721, -80.5853]        # Example highway coord

# Calculate distances
distance_city = haversine(launch_site, city)
distance_railway = haversine(launch_site, railway)
distance_highway = haversine(launch_site, highway)

print(f"Distance to nearest city: {distance_city:.2f} km")
print(f"Distance to nearest railway: {distance_railway:.2f} km")
print(f"Distance to nearest highway: {distance_highway:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
